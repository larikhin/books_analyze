## подготовительная работа

```python
# Создаем функции для создания датасета
# Функция загружает книгу в формате .txt, записывают все слова книги в список
# затем возвращает кортеж, его передает ее в функцию записи кортежа в датафрейм 
import pandas as pd


# функция читает текстовый фаил и создает кортеж из слов файла
def from_file_to_tuple(bookname):
    filename = './txt/' + bookname + '.txt'
    # открываем фаил
    with open(filename, 'r') as f:
        list_word=[]
        # запись предложений текста в список     
        lines = f.readlines()
        # цикл для каждой строки
        for line in lines:
            # нарезка предложения на отдельные слова
            words = line.strip().split()
            for word in words:
                # удаление мусора из слов
                word = word.strip('--––<>.»%\/* [„=,  r;№«{}[]()!?&:"1234567890')
                word = word.strip(" ' ")
                word = word.strip("—")
                word = word.strip('…“ ')
                word = word.strip('" ')
                words = line.strip('\t/t')
                # единый регистр всех слов
                word=word.lower()
                word = word.strip('1234567890abcdefghijklnmopqrstuvwzyx')
                word=word.replace("à", "а")
                word=word.replace('а́','а')
                word=word.replace('ѐ','е')
                # добавление слова в список
                list_word.append(word)
        # удаление пустых строк 
        list_word = ' '.join(list_word).split()
        # перевод в кортеж
        tuple_word = tuple(list_word)
        # возвращаем кортеж всех слов книги и закрываем фаил
        return tuple_word
 
def to_df(bookname):
    # вызов функции чтения файла, создания кортежа слов
    tuple_word = from_file_to_tuple(bookname)
    # создание датафрейма
    table_words = pd.DataFrame({bookname:tuple_word})
    return table_words
```
